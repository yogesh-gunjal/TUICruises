# Reading Dataset

```python
import pandas as pd

# URL to the raw CSV file from GitHub
url = 'https://raw.githubusercontent.com/yogesh-gunjal/TUICruises/refs/heads/main/task_data/data.csv'

# Read the CSV file directly from GitHub
fleet_data = pd.read_csv(url)

# Display the first few rows of the dataset
display(fleet_data.head(5))
```

# Data Pre-processing
5 minutes to hourly interval conversion

```python
# Remove minutes and seconds, keep only date and hour
fleet_data['Start Time'] = fleet_data['Start Time'].str[:13]
# Drop the 'End Time' column
fleet_data = fleet_data.drop(columns=['End Time'])
# Group by 'Start Time' and 'Vessel Name', then sum other columns
fleet_data_hourly = fleet_data.groupby(['Start Time', 'Vessel Name']).sum().reset_index()
display(fleet_data_hourly)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ... (previous code to create 'fleet_data_hourly')

# Convert 'Start Time' back to datetime for plotting
fleet_data_hourly['Start Time'] = pd.to_datetime(fleet_data_hourly['Start Time'])

# Function to create line chart for a specific galley
def create_line_chart(galley_column, title):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6), sharey=True)  # 1 row, 2 columns
    fig.suptitle(title)  # Overall title for the figure

    # Vessel 1
    vessel1_data = fleet_data_hourly[fleet_data_hourly['Vessel Name'] == 'Vessel 1']
    ax1.plot(vessel1_data['Start Time'], vessel1_data[galley_column], label='Vessel 1')
    ax1.set_xlabel('Start Time')
    ax1.set_ylabel(galley_column)
    ax1.set_title('Vessel 1')
    ax1.legend()
    plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')

    # Vessel 2
    vessel2_data = fleet_data_hourly[fleet_data_hourly['Vessel Name'] == 'Vessel 2']
    ax2.plot(vessel2_data['Start Time'], vessel2_data[galley_column], label='Vessel 2', color='orange')
    ax2.set_xlabel('Start Time')
    ax2.set_title('Vessel 2')
    ax2.legend()
    plt.setp(ax2.get_xticklabels(), rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

# Create chart for Power Galley 1
create_line_chart('Power Galley 1 (MW)', 'Power Galley 1 (MW) over Time')

# Create chart for Power Galley 2
create_line_chart('Power Galley 2 (MW)', 'Power Galley 2 (MW) over Time')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ... (previous code to create 'fleet_data_hourly')

# Convert 'Start Time' back to datetime for plotting and filtering
fleet_data_hourly['Start Time'] = pd.to_datetime(fleet_data_hourly['Start Time'])

# Filter data for Vessel 2 from July to September
vessel2_data_filtered = fleet_data_hourly[
    (fleet_data_hourly['Vessel Name'] == 'Vessel 2') &
    (fleet_data_hourly['Start Time'].dt.month.between(7, 8))  # July to September
]

# Function to create line chart for a specific galley (only for Vessel 2)
def create_line_chart(galley_column, title):
    fig, ax = plt.subplots(figsize=(12, 6))  # Only one subplot now
    fig.suptitle(title)

    # Vessel 2 (filtered data)
    ax.plot(vessel2_data_filtered['Start Time'], vessel2_data_filtered[galley_column], label='Vessel 2', color='orange')
    ax.set_xlabel('Start Time')
    ax.set_ylabel(galley_column)
    ax.set_title('Vessel 2 (July-Sept)')  # Updated title
    ax.legend()
    plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

# Create chart for Power Galley 1
create_line_chart('Power Galley 1 (MW)', 'Power Galley 1 (MW) over Time (Vessel 2, July-Sept)')

# Create chart for Power Galley 2
create_line_chart('Power Galley 2 (MW)', 'Power Galley 2 (MW) over Time (Vessel 2, July-Sept)')
```

```python
from google.colab import files
# Define the columns related to power consumption
power_columns = [
    'Power Galley 1 (MW)', 'Power Galley 2 (MW)', 'Power Service (MW)',
    'HVAC Chiller 1 Power (MW)', 'HVAC Chiller 2 Power (MW)', 'HVAC Chiller 3 Power (MW)',
    'Scrubber Power (MW)', 'Propulsion Power (MW)', 'Port Side Propulsion Power (MW)',
    'Starboard Side Propulsion Power (MW)', 'Bow Thruster 1 Power (MW)', 'Bow Thruster 2 Power (MW)',
    'Bow Thruster 3 Power (MW)', 'Stern Thruster 1 Power (MW)', 'Stern Thruster 2 Power (MW)'
]

# Sum the power consumption columns to create the 'Total Power Consumption' column
fleet_data_hourly['Total Power Consumption (MW)'] = fleet_data_hourly[power_columns].sum(axis=1)

# Select oly the 'Start Time' and 'Total Power Consumption (MW)' columns with other consumption columns
Power_consumption_summary = fleet_data_hourly[['Start Time', 'Vessel Name','Total Power Consumption (MW)','Power Galley 1 (MW)', 'Power Galley 2 (MW)', 'Power Service (MW)',
    'HVAC Chiller 1 Power (MW)', 'HVAC Chiller 2 Power (MW)', 'HVAC Chiller 3 Power (MW)',
    'Scrubber Power (MW)', 'Propulsion Power (MW)', 'Port Side Propulsion Power (MW)',
    'Starboard Side Propulsion Power (MW)', 'Bow Thruster 1 Power (MW)', 'Bow Thruster 2 Power (MW)',
    'Bow Thruster 3 Power (MW)', 'Stern Thruster 1 Power (MW)', 'Stern Thruster 2 Power (MW)']]

# Display the summary dataframe
display(Power_consumption_summary)

Power_consumption_summary.to_csv('Power_consumption_summary.csv', index=False)
files.download('Power_consumption_summary.csv')
```

# Correlation Matrix
Correlation Between Variables

Identify strong correlations between variables

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for correlation calculation
numerical_columns = fleet_data.select_dtypes(include=['number']).columns
correlation_matrix = fleet_data[numerical_columns].corr()

# Create a heatmap
plt.figure(figsize=(12, 10))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Define the columns for fuel flow rate and diesel generator fuel consumption
fuel_flow_columns = ['Main Engine 1 Fuel Flow Rate (kg/h)', 'Main Engine 2 Fuel Flow Rate (kg/h)',
                       'Main Engine 3 Fuel Flow Rate (kg/h)', 'Main Engine 4 Fuel Flow Rate (kg/h)']
diesel_gen_columns = ['Diesel Generator 1 Power (MW)', 'Diesel Generator 2 Power (MW)',
                             'Diesel Generator 3 Power (MW)', 'Diesel Generator 4 Power (MW)']

# Create a subset of the DataFrame with these columns
fuel_data = fleet_data[[*fuel_flow_columns, *diesel_gen_columns]]

# Calculate the correlation matrix
correlation_matrix = fuel_data.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Fuel Flow Rate vs. Diesel Generator Fuel Consumption')
plt.show()
```

```python
# Define the columns for Main Engine Fuel Flow Rates
main_engine_fuel_flow_columns = ['Main Engine 1 Fuel Flow Rate (kg/h)', 'Main Engine 2 Fuel Flow Rate (kg/h)',
                                    'Main Engine 3 Fuel Flow Rate (kg/h)', 'Main Engine 4 Fuel Flow Rate (kg/h)']

# Calculate the total fuel flow for each hour
fleet_data_hourly['Total Main Engine Fuel Flow (kg/h)'] = fleet_data_hourly[main_engine_fuel_flow_columns].sum(axis=1)

# Create a new DataFrame with the desired columns
main_engine_fuel_flow_hourly = fleet_data_hourly[['Start Time','Vessel Name', 'Total Main Engine Fuel Flow (kg/h)',
                                                    *main_engine_fuel_flow_columns]]  # Include individual fuel flow rates

# Display the new DataFrame
display(main_engine_fuel_flow_hourly)

main_engine_fuel_flow_hourly.to_csv('Engine_Fuel_summary.csv', index=False)
files.download('Engine_Fuel_summary.csv')
```
